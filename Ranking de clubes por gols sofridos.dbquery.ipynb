```sql
SELECT
    c.clube,
    COUNT(*) AS gols_sofridos
FROM fato_partida fp
JOIN dim_clube c ON c.clube_id IN (fp.mandante_id, fp.visitante_id)
JOIN fato_gol fg ON fg.partida_id = fp.partida_id
WHERE 
    (c.clube_id = fp.mandante_id AND fg.clube_id = fp.visitante_id)
    OR
    (c.clube_id = fp.visitante_id AND fg.clube_id = fp.mandante_id)
GROUP BY c.clube
ORDER BY gols_sofridos DESC;
```
